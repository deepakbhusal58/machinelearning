import all the libraries needed for the analysis


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import seaborn as sns
```

```python
# dataset is loaded into the data variable
data = pd.read_csv('D:\datasets\housing.csv')
data.head(100)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# all the rows with null values are removed
data.dropna(inplace=True)
```

```python
data['ocean_proximity'].value_counts()
```

ocean_proximity's label is divided into 5 features in a dataset


```python
# the 5 features of ocean_proximity is concatenate with the dataset b removing ocean_proximity itself
train_data=data.join(pd.get_dummies(data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
```

```python
train_data
```

```python
# train_test_split is imported to divide the dataset for training and spliting data

from sklearn.model_selection import train_test_split

X = train_data.drop(['median_house_value'], axis=1)
y = train_data['median_house_value'] 
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# the model is trained 
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

reg.fit(X_train,y_train)
```

```python
X_test
```

```python
reg.score(X_test, y_test)
print("Linear regression Model", reg.score(X_test, y_test))
```
